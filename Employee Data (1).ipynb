```python
# There is a duplicate employee in 2068

employee_data.sort_values("EmployeeNumber", inplace=True)
employee_data.drop_duplicates(subset="EmployeeNumber",
                     keep=False, inplace=True)
```

```python
# There is a incomplete data that we need to interpolate - we figure out which column it is.


employee_data.isna().any()
```

```python
# We discover it is 'Over18'. We then determine if 'Over18' is Y, from the age and then interpolate the NaN values, and keep performing this on a looop until the blanks are empty

employee_data.loc[employee_data['Age'] >= 18, 'Over18'] = 'Y'
employee_data.loc[employee_data['Age'] < 18, 'Over18'] = 'N'


print(employee_data['Over18'].head(12))
```

```python
# We want to categorise the ages into bands, and enter a new column 'age_category'
employee_data.loc[employee_data['Age'] >= 65,'AgeCategory'] = '65+'
employee_data.loc[employee_data['Age'] <= 64,'AgeCategory'] = '45-64'
employee_data.loc[employee_data['Age'] <= 44,'AgeCategory'] = '35-44'
employee_data.loc[employee_data['Age'] <= 34,'AgeCategory'] = '25-34'
employee_data.loc[employee_data['Age'] <= 24,'AgeCategory'] = '15-24'
print(employee_data.head(3))
```

```python
# We want to categorise the Job Satisfaction into broader categories
employee_data.loc[employee_data['JobSatisfaction'] <= 2,'EnjoyJob'] = 'N'
employee_data.loc[employee_data['JobSatisfaction'] >= 3,'EnjoyJob'] = 'Y'
print(employee_data.head(3))
```

```python
# We want to compare the job satisfaction and of IBM employees to a benchmark. We also wish to benchmark IBM employees of their worrk/life balance. 
# We therefore incorporate another dataset, an open-source Gallup Global Wellbeing Index and focus in on Balance and Satisfaction.
# We wish to account for any variations by Gender as well as Age Group
from urllib.request import urlretrieve
url = 'https://www.gallup.com/file/education/468923/Global%20Wellbeing%20Initiative%20Dataset.xlsx'
global_wellness_data = urlretrieve(url, 'Global%20Wellbeing%20Initiative%20Dataset.xlsx')
print(global_wellness_data)
```

```python
# Create a pivot table to roll up Job Satisfaction into broader categories, we need to keep the margins along the side to calculate the %

import numpy as np
import pandas as pd
table = employee_data.pivot_table(values="JobSatisfaction", index="AgeCategory", columns="EnjoyJob", margins=True, fill_value=0, aggfunc="count") 
table_margin = table.div(table.iloc[:,-1], axis=0).mul(100).astype (str) + '%'
piv = table_margin.drop(index='All', axis=0)
print(round(piv,2))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.lineplot(x='Y', y='AgeCategory', data=piv)
plt.show()
# Showing the slight but steadily decline in job satisfaction over the age categories
```
